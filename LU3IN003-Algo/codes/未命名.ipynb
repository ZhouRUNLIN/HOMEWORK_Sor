```python
from projet import *
import psutil

def prog_dyn(x, y):
    """
        renvoie Ã  la fois la distance d(x, y) et un alignement optimal
    """
    T=create_T(x, y)
    print('distance de ' + x + ' et ' + y + ' : ' + (str)(dist_1(x, y)))
    print('alignement optimal de' + x + ' et ' + y + ' : ' + (str)(sol_1(T, x, y)))

#pre-Test : pour les mots avec la taille plus grand que xx, le temp d'execution est plus que 10min

listeFic={ 0: "Instances_genome/Inst_0000010_44.adn",
           1: "Instances_genome/Inst_0000010_8.adn",
           2: "Instances_genome/Inst_0000010_7.adn",
           3: "Instances_genome/Inst_0000012_13.adn",
           4: "Instances_genome/Inst_0000012_56.adn",
           5: "Instances_genome/Inst_0000013_45.adn"
         }
```

```python
tailleMot=[5, 9, 10, 11, 12, 13]
cpuPercent=[]

for i in range(0, len(listeFic)):
    t0=time.time()
    cp=read_file(listeFic[i])
    cpu=max(psutil.cpu_percent(interval=5, percpu=True))
    prog_dyn(cp[2], cp[3])
    cpuPercent.append(cpu)
    t1=time.time()
    print("time taken for execute prog_dyn :" + (str)(t1-t0) + "\n")

print(tailleMot)
print(cpuPercent)               
```
